```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
correlation = X[X.columns.difference(["reason for choosing their major", "major"])].apply(lambda x: x.corr(y))

plt.figure(figsize=(10, 6))
correlation.plot(kind='bar')
plt.title('Correlation of Features with GPA')
plt.axhline(0, color='red')  # Draw a horizontal line at y=0
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
X = pd.get_dummies(X, columns=["reason for choosing their major", "major"], dtype="int64")
```

```python
X
```

```python
model = LinearRegression()
model.fit(X, y)
```
